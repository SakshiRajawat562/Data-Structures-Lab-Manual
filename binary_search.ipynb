```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1  # element not found


# New Example
arr = [5, 12, 19, 27, 33, 45, 52, 60]
target = 33

result = binary_search(arr, target)
print("Element found at index:", result)
```
