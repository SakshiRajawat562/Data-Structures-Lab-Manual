```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i   # return index
    return -1  # not found


# New Example
arr = [5, 12, 7, 25, 18, 30, 42]
target = 18

result = linear_search(arr, target)
print("Element found at index:", result)
```
