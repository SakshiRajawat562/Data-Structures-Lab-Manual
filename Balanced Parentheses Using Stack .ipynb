```python
def is_balanced(expression):
    stack = []
    opening = "({["
    closing = ")}]"
    matching = {')': '(', '}': '{', ']': '['}

    for char in expression:
        if char in opening:
            stack.append(char)
        elif char in closing:
            if not stack or stack[-1] != matching[char]:
                return False
            stack.pop()

    return len(stack) == 0


# Example test cases
print(is_balanced("{[()]}"))     
print(is_balanced("{[(])}"))     
print(is_balanced("((()))"))     
print(is_balanced("((()"))       
```
